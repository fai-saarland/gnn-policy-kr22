```python
import json
import matplotlib.pyplot as plt
```

```python
directory_path = "/Users/nicola_mueller/Desktop/general_policies/lightning_logs/blocks_retrain/version_0"
train_losses_path = directory_path + "/losses.train"
bug_losses_path = directory_path + "/losses.bugs"

with open(train_losses_path, "r") as f:
    train_losses = json.load(f)
with open(bug_losses_path, "r") as f:
    bug_losses = json.load(f)

plt.title("Train Losses vs. Bug Losses")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.plot(train_losses, label="train")
plt.plot(bug_losses, label="bug")
#print(bug_losses)
plt.legend()
```
